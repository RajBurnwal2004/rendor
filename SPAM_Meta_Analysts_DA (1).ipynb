```python
categorical_cols=df.iloc[:,[0,1,2,3,4,5,7,8,10,11]]
from scipy.stats import f_oneway

anova_results = {}
for col in categorical_cols:
    groups = [df[df[col] == val]['Price'] for val in df[col].unique()]
    f_stat, p_value = f_oneway(*groups)
    anova_results[col] = p_value

anova_results = {k: v for k, v in sorted(anova_results.items(), key=lambda item: item[1])}

for feature, p_val in anova_results.items():
    print(f"Feature: {feature}, p-value: {p_val}")
```

```python
percentile_25 = new_df['Price'].quantile(0.25)
percentile_75 = new_df['Price'].quantile(0.75)
iqr = percentile_75 - percentile_25
lower_bound = percentile_25 - 1.5 * iqr
upper_bound = percentile_75 + 1.5 * iqr
```

```python
df=new_df.copy()
df['Price']=np.where(
    df['Price'] > upper_bound,
    upper_bound,
    np.where(
        df['Price'] < lower_bound,
        lower_bound,
        df['Price']
    )
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.iloc[:,:],df.iloc[:,6],test_size=0.3,random_state=0)
X_train.drop('Price',axis=1,inplace=True)
X_test.drop('Price',axis=1,inplace=True)
```

```python
X_train.sample(10)
```

```python
sorted_case_materials
```

```python
sorted_materials
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe_mov=OrdinalEncoder(categories=[['Solar','Smartwatch','Quartz','Automatic','Manual winding']])
oe_mov.fit(X_train[['Movement']])
X_train[['Movement']]=oe_mov.fit_transform(X_train[['Movement']])
X_test[['Movement']]=oe_mov.fit_transform(X_test[['Movement']])

oe_cas=OrdinalEncoder(categories=[sorted_case_materials])
oe_cas.fit(X_train[['Case material']])
X_train[['Case material']]=oe_cas.transform(X_train[['Case material']])
X_test[['Case material']]=oe_cas.transform(X_test[['Case material']])

oe_con=OrdinalEncoder(categories=[['Used (Incomplete)','Used (Poor)','Used (Fair)','Used (Good)','New','Like new & unworn','Used (Very good)']])
oe_con.fit(X_train[['Condition']])
X_train[['Condition']]=oe_con.transform(X_train[['Condition']])
X_test[['Condition']]=oe_con.transform(X_test[['Condition']])

oe_sc=OrdinalEncoder(categories=[['No original box, no original papers','Original box, no original papers','Original papers, no original box','Original box, original papers']])
oe_sc.fit(X_train[['Scope of delivery']])
X_train[['Scope of delivery']]=oe_sc.transform(X_train[['Scope of delivery']])
X_test[['Scope of delivery']]=oe_sc.transform(X_test[['Scope of delivery']])

oe_gen=OrdinalEncoder(categories=[["Women's watch","Men's watch/Unisex"]])
oe_gen.fit(X_train[['Gender']])
X_train[['Gender']]=oe_gen.transform(X_train[['Gender']])
X_test[['Gender']]=oe_gen.transform(X_test[['Gender']])

oe_ava=OrdinalEncoder(categories=[['Item is in stock','Item available on request','Item needs to be procured']])
oe_ava.fit(X_train[['Availability']])
X_train[['Availability']]=oe_ava.transform(X_train[['Availability']])
X_test[['Availability']]=oe_ava.transform(X_test[['Availability']])

oe_cr=OrdinalEncoder(categories=[['Mineral Glass','Plastic','Glass','Plexiglass','Sapphire crystal']])
oe_cr.fit(X_train[['Crystal']])
X_train[['Crystal']]=oe_cr.transform(X_train[['Crystal']])
X_test[['Crystal']]=oe_cr.transform(X_test[['Crystal']])

oe_cl=OrdinalEncoder(categories=[['Fold clasp, hidden','Buckle','Jewelry clasp','Fold clasp','Double-fold clasp','No clasp']])
oe_cl.fit(X_train[['Clasp']])
X_train[['Clasp']]=oe_cl.transform(X_train[['Clasp']])
X_test[['Clasp']]=oe_cl.transform(X_test[['Clasp']])

oe_bra=OrdinalEncoder(categories=[sorted_materials])
oe_bra.fit(X_train[['Bracelet material']])
X_train[['Bracelet material']]=oe_bra.transform(X_train[['Bracelet material']])
X_test[['Bracelet material']]=oe_bra.transform(X_test[['Bracelet material']])
```

```python
X_train.sample(10)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=pd.DataFrame(sc.transform(X_train),columns=X_train.columns)
X_test=pd.DataFrame(sc.transform(X_test),columns=X_test.columns)


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y_train_scaled = scaler.fit_transform(Y_train.values.reshape(-1, 1)).ravel()
y_test_scaled = scaler.transform(Y_test.values.reshape(-1, 1)).ravel()
Y_train=pd.DataFrame(y_train_scaled,columns=['Price'])
Y_test=pd.DataFrame(y_test_scaled,columns=['Price'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
rf_regressor = RandomForestRegressor(max_features='sqrt',min_samples_leaf=4,min_samples_split=10,max_depth=10,bootstrap=True,n_estimators=500,random_state=42)
rf_regressor.fit(X_train, Y_train.values.ravel())
y_pred = rf_regressor.predict(X_test)
mse_rf = mean_squared_error(Y_test, y_pred)
print("Random Forest Mean Squared Error: ",mse_rf)
r2_score_value = r2_score(Y_test, y_pred)
print('Random Forest r2 score:',r2_score_value)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(max_features='sqrt',min_samples_leaf=4,min_samples_split=10,max_depth=10,random_state=42)
dt_regressor.fit(X_train, Y_train)
y_pred_dt = dt_regressor.predict(X_test)
mse_dt = mean_squared_error(Y_test, y_pred_dt)
print("Decision Tree Regressor Mean Squared Error: ",mse_dt)
r2_score_value = r2_score(Y_test, y_pred_dt)
print('Decision Tree r2 score:',r2_score_value)
```

```python
clf1=RandomForestRegressor(random_state=42)
clf2=DecisionTreeRegressor(max_depth=10,random_state=42)
estimators=[('rf_regressor',clf1),('dt_regressor',clf2)]
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
vc=VotingRegressor(estimators=estimators)
for estimator in estimators:
  x=cross_val_score(estimator[1],X_train,Y_train.values.ravel(),cv=10,scoring='r2')
  print(estimator[0],np.round(np.mean(x),2))
vc.fit(X_train,Y_train.values.ravel())
y_pred_vc=vc.predict(X_test)
r2_score_value = r2_score(Y_test, y_pred_vc)
print('Voting r2 score:',r2_score_value)
mse_vc = mean_squared_error(Y_test, y_pred_vc)
print("Voting Mean Squared Error: ",mse_vc)
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
train_data = lgb.Dataset(X_train, label=Y_train.values.ravel())
valid_data = lgb.Dataset(X_test, label=Y_test.values.ravel(), reference=train_data)
params = {
    'objective': 'regression',
    'metric': 'mse',    'learning_rate': 0.1,
    'num_leaves': 51,
    'random_state': 42
}
lgb_model = lgb.train(
    params,
    train_data,
    valid_sets=[valid_data],
    num_boost_round=300
    )
y_pred = lgb_model.predict(X_test)
mse = mean_squared_error(Y_test.values.ravel(), y_pred)
print("LightGBM MSE:", mse)
r2_score_value = r2_score(Y_test, y_pred)
print('LightGBM r2 score:',r2_score_value)
```
